## Get total contributions from top 5 contributing individuals, for all recipients, in all of Oregon

[Github](https://github.com/hackoregon/btc-backend/blob/master/endpoints/endpoint_use_readme.md#get-total-contributions-from-top-5-contributing-individuals-for-all-recipients-in-all-of-oregon)

```python
# get the data
tic_file = 'http://54.213.83.132/hackoregon/http/oregon_individual_contributors/5/'
tic_data = pd.read_json(tic_file)
df = pd.DataFrame(tic_data, columns=['contributor_payee', 'sum'])
df.index = df['contributor_payee']
df.plot(kind='bar',figsize=(12,6))
plt.title('Top 5 Contributing Individuals')
plt.show()

# John Arnold contributed $2.75mm
```

## All oregon summary

[Github](https://github.com/hackoregon/btc-backend/blob/master/endpoints/endpoint_use_readme.md#all-oregon-summary)

```python
aos_file = 'http://54.213.83.132/hackoregon/http/all_oregon_sum/_/'
aos_data = pd.read_json(aos_file)
df = pd.DataFrame(aos_data, columns=['from_outside', 'from_within', 'in', 'out', 'to_outside', 'to_within', 'total_from_in_state', 'total_grass_roots'])
df.plot(kind='barh',figsize=(12,6))
plt.title('Summary of Oregon political transactions')
plt.show()
```

**Definitions**
- **from_outside**: money coming from outside the political system
- **from_within**: money coming from entities within the political system
- **in**: money recieved by all sources
- **out**: money transfered from all sources
- **to_outside**: money going to entities outside the political system
- **to_within**: money transfered to entities within the political system
- **total_from_in_state**: total donations coming from withing Oregon
- **total_grass_roots**: total money from grass roots donations (those less than $200)

```python
df = pd.DataFrame(aos_data, columns=['from_outside', 'from_within'])
df.plot(kind='barh')
plt.title('Most money comes from outside \n the political system')
plt.show()
```

```python
df = pd.DataFrame(aos_data, columns=['in', 'out'])
df.plot(kind='barh')
plt.title('Slightly more money is recieved than transferred \n out of the state political system')
plt.show()
```

```python
df = pd.DataFrame(aos_data, columns=['to_outside', 'from_within'])
df.plot(kind='barh')
plt.title('Most money is transferred outside \n the political system (advertisting, etc.)')
plt.show()
```
