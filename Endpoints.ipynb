```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
matplotlib.style.use('ggplot')
%matplotlib inline
```

## Get total contributions from top 5 contributing individuals, for all recipients, in all of Oregon

[Github](https://github.com/hackoregon/btc-backend/blob/master/endpoints/endpoint_use_readme.md#get-total-contributions-from-top-5-contributing-individuals-for-all-recipients-in-all-of-oregon)

```python
# get the data
tic_file = 'http://54.213.83.132/hackoregon/http/oregon_individual_contributors/5/'
tic_data = pd.read_json(tic_file)
df = pd.DataFrame(tic_data, columns=['contributor_payee', 'sum'])
df.index = df['contributor_payee']
df.plot(kind='bar')
plt.show()

# John Arnold contributed $2.75mm
```

## All oregon summary

[Github](https://github.com/hackoregon/btc-backend/blob/master/endpoints/endpoint_use_readme.md#all-oregon-summary)

```python
aos_file = 'http://54.213.83.132/hackoregon/http/all_oregon_sum/_/'
aos_data = pd.read_json(aos_file)
df = pd.DataFrame(aos_data, columns=['from_outside', 'from_within', 'in', 'out', 'to_outside', 'to_within', 'total_from_in_state', 'total_grass_roots'])
df.plot(kind='barh')
plt.show()
```

```python
# Most money comes from outside the political system

df = pd.DataFrame(aos_data, columns=['from_outside', 'from_within'])
df.plot(kind='barh')
plt.show()
```

```python
# Slightly more money is recieved than transferred out of the state political system.

df = pd.DataFrame(aos_data, columns=['in', 'out'])
df.plot(kind='barh')
plt.show()
```

```python
# Most money is transferred outside the political system (advertisting, etc.)

df = pd.DataFrame(aos_data, columns=['to_outside', 'from_within'])
df.plot(kind='barh')
plt.show()
```
